# outlayers with IQR

```python
sns.boxplot(data=numerical_col)
plt.xticks(rotation=90)
```

```python
def wisker(col):
  q1=numerical_col[col].quantile(0.25)
  q3=numerical_col[col].quantile(0.75)
  iqr=q3-q1
  lower=q1-(1.5*iqr)
  upper=q3+(1.5*iqr)
  return lower,upper
```

```python
for i in numerical_col.select_dtypes(include=['number']).columns:
 lower,upper=wisker(i)
 numerical_col[i]=np.where(numerical_col[i]<lower,lower,numerical_col[i])   
 numerical_col[i]=np.where(numerical_col[i]>upper,upper,numerical_col[i])
```

```python
import mlflow
import mlflow.sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.decomposition import PCA
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix, accuracy_score

# Assuming x_train, x_test, y_train, y_test are already defined

# Number of principal components (you can adjust this based on your data)
n_components = 10

# PCA transformation
pca = PCA(n_components=n_components)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.transform(x_test)

# Logging the explained variance ratio
explained_variance = pca.explained_variance_ratio_

with mlflow.start_run():
    # Log PCA parameters
    mlflow.log_param('pca_n_components', n_components)
    for i, variance in enumerate(explained_variance):
        mlflow.log_metric(f'pca_explained_variance_{i}', variance)
    
    # Model initialization
    LogReg = LogisticRegression(max_iter=1000)
    RandFor = RandomForestClassifier(n_estimators=500, max_depth=6)
    GBM = GradientBoostingClassifier(n_estimators=300, max_depth=6, learning_rate=0.01, min_samples_split=5)

    # Logging parameters
    mlflow.log_param('logreg_max_iter', 1000)
    mlflow.log_param('rf_n_estimators', 500)
    mlflow.log_param('rf_max_depth', 6)
    mlflow.log_param('gbm_n_estimators', 300)
    mlflow.log_param('gbm_max_depth', 6)
    mlflow.log_param('gbm_learning_rate', 0.01)
    mlflow.log_param('gbm_minimum_samples_split', 5)

    # Training models
    lr = LogReg.fit(x_train_pca, y_train)
    rf = RandFor.fit(x_train_pca, y_train)
    gbm = GBM.fit(x_train_pca, y_train)

    # Predictions
    y_pred_lr = lr.predict(x_test_pca)
    y_pred_rf = rf.predict(x_test_pca)
    y_pred_gbm = gbm.predict(x_test_pca)

    # Evaluation metrics
    accuracy_lr = accuracy_score(y_test, y_pred_lr)
    precision_lr = precision_score(y_test, y_pred_lr, average='weighted')
    recall_lr = recall_score(y_test, y_pred_lr, average='weighted')
    f1_lr = f1_score(y_test, y_pred_lr, average='weighted')
    cm_lr = confusion_matrix(y_test, y_pred_lr)

    accuracy_rf = accuracy_score(y_test, y_pred_rf)
    precision_rf = precision_score(y_test, y_pred_rf, average='weighted')
    recall_rf = recall_score(y_test, y_pred_rf, average='weighted')
    f1_rf = f1_score(y_test, y_pred_rf, average='weighted')
    cm_rf = confusion_matrix(y_test, y_pred_rf)

    accuracy_gbm = accuracy_score(y_test, y_pred_gbm)
    precision_gbm = precision_score(y_test, y_pred_gbm, average='weighted')
    recall_gbm = recall_score(y_test, y_pred_gbm, average='weighted')
    f1_gbm = f1_score(y_test, y_pred_gbm, average='weighted')
    cm_gbm = confusion_matrix(y_test, y_pred_gbm)

    # Logging metrics
    mlflow.log_metric("lr_accuracy", accuracy_lr)
    mlflow.log_metric("lr_precision", precision_lr)
    mlflow.log_metric("lr_recall", recall_lr)
    mlflow.log_metric("lr_f1", f1_lr)
    mlflow.log_metric("rf_accuracy", accuracy_rf)
    mlflow.log_metric("rf_precision", precision_rf)
    mlflow.log_metric("rf_recall", recall_rf)
    mlflow.log_metric("rf_f1", f1_rf)
    mlflow.log_metric("gbm_accuracy", accuracy_gbm)
    mlflow.log_metric("gbm_precision", precision_gbm)
    mlflow.log_metric("gbm_recall", recall_gbm)
    mlflow.log_metric("gbm_f1", f1_gbm)

    # Logging models
    mlflow.sklearn.log_model(lr, "logistic-regression-model")
    mlflow.sklearn.log_model(rf, "random-forest-model")
    mlflow.sklearn.log_model(gbm, "gradient-boosting-model")

    print("Confusion Matrix for Logistic Regression:")
    print(cm_lr)
    print("Confusion Matrix for Random Forest:")
    print(cm_rf)
    print("Confusion Matrix for Gradient Boosting:")
    print(cm_gbm)
```
